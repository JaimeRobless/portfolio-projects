# Coffee Data Exploration

##### What factors determine the sale of a specific coffee?

## Goal

The goal of this project is to identify the factors that contribute to the sale of coffee,specifically sales by coffee type,sales by hour of the date or revenue trends over time.By performing data analysis using python, this project aims to provide valuable insight that can help can help this hypothetical Coffee Shop make more informed decisions about what type of coffee to promote during a specific time or what they can cut cost.

## Cleaning the Data

```python
# Looping through the data and find if anything is missing 

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(percent_missing * 100)))
```

#### Sales by Coffee Type Depending on The Month

```python
coffee_count = pd.crosstab(df["Month_name"],df["coffee_name"])
coffee_count = coffee_count.reindex(index=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
coffee_count.plot.bar(title="Coffee Type of Sold Depending on month",ylabel="Number of Coffee Sold",xlabel="Month")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',title="Coffee Type")
```

```python
coffee_ptcs = coffee_count.div(coffee_count.sum(axis="columns"),axis="index")
coffee_ptcs.plot.bar(stacked=True,title="Percentage of Coffee Types depending on the month",ylabel="Percentage",xlabel="Month")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',title="Coffee Type")
```

#### Sales by Time the Day

```python
time_day = pd.crosstab(df["Time_of_Day"],df["coffee_name"])
time_day = time_day.reindex(index=["Morning","Afternoon","Night"])
time_day.plot.bar(title="Coffee Sales depending on the Time of Day",ylabel="Number of Coffee Sold",xlabel="Time of Day")
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left',title="Coffee Type")
```

```python
time_ptcs = time_day.div(time_day.sum(axis="columns"),axis="index")
time_ptcs.plot.bar(stacked=True,title="Percentage of Coffee Types depending on Time of Day",ylabel="Percentage",xlabel="Time of Day")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left',title="Coffee Type")
```
